```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://myanimelist.net/topmanga.php"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
```

```python
manga_list = []

for manga in soup.find_all("tr", "ranking-list"):
    title = manga.find("a", class_="hoverinfo_trigger fs14 fw-b")
    title_tag = title.text
    link = title["href"]
    manga_list.append({"Title": title_tag, "Link": link})
```

```python
def scrape_manga_reviews(manga_url):
    manga_page = requests.get(manga_url)
    soup = BeautifulSoup(manga_page.text, "html.parser")

    rating = soup.find("span", itemprop="ratingValue").text
    desc = soup.find("span", itemprop="description").text.strip()
    
    return rating, desc
```

```python
import time
data = []

for manga in manga_list:
    title = manga["Title"]
    link = manga["Link"]

    try:
        rating, desc = scrape_manga_reviews(link)
        data.append({"Title": title, "Link": link, "Rating": rating, "Description": desc})
    except Exception as e:
        print(f"Failed to scrape {title}: {e}")

    time.sleep(2)
```

```python
import pandas as pd

df = pd.DataFrame(data)

df.head()
```
